```python
import streamlit as st
import pandas as pd
import numpy as np
import chardet
import base64
from sklearn.metrics.pairwise import cosine_similarity
```

```python
with open('NetflixOriginals.csv', 'rb') as f:
    result = chardet.detect(f.read())
    encoding = result['encoding']


movies_df = pd.read_csv('NetflixOriginals.csv', encoding=encoding)
```

```python
def get_recommendations(title, cosine_sim, movies_df):
    # Get the index of the movie that matches the title
    idx = movies_df[movies_df['Title'] == title].index[0]

    # Get the pairwise similarity scores with all movies
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the movies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the indices of the top 10 most similar movies
    movie_indices = [i[0] for i in sim_scores[1:11]]

    # Return the top 10 most similar movies
    return movies_df['Title'].iloc[movie_indices]

with open('NetflixOriginals.csv', 'rb') as f:
    result = chardet.detect(f.read())
    encoding = result['encoding']

# Load the movie dataset
df = pd.read_csv('NetflixOriginals.csv', encoding=encoding)

# Preprocess the data
df.dropna(inplace=True)
df = df[df['Genre'].apply(lambda x: x.isnumeric() == False)]

# Get the unique genres in the dataset
genres = set(x for l in df['Genre'].str.split("|").tolist() for x in l)

# Create a binary feature for each genre
for genre in genres:
    df[genre] = df['Genre'].str.contains(genre).astype(int)

# Select the features for the model
movies_features = df[genres]

# Calculate the cosine similarity matrix
cosine_sim = cosine_similarity(movies_features)

# Create the Streamlit UI

with open("netflix.jpeg", "rb") as image_file:
    encoded_string = base64.b64encode(image_file.read())
st.markdown(
f"""    
<style>
.stApp {{
    background-image: url(data:image/{"jpeg"};base64,{encoded_string.decode()});
    background-size: cover
}}
</style>
""",
unsafe_allow_html=True
)
 
selected_movie = st.selectbox("A movie that you saw recently".format(color='white'), df['Title'].tolist(), key='select_movie')

if st.button("Recommend Movie"):
    # Get the recommended movies
    recommendations = get_recommendations(selected_movie, cosine_sim, df)
    
    # Play the video
    video_file = "F:\\Simple_projects\\Netflix\\Netflix New Logo Animation 2019.mp4"
    st.video(video_file)

    # Display the recommended movies
    st.markdown("<p style='color: white; font-size: 24px; font-weight: bold;'>Search Results:</p>", unsafe_allow_html=True)
    for movie in recommendations:
        st.markdown("<p style='color: white;'>- " + movie + "</p>", unsafe_allow_html=True)

```
